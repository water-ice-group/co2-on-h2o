```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
# Density - CO2

dat_BLYP = dict()
counter = 1
step = 1
# ---------------------------------------------------------------------
dir = './low_pressures/'
for i in range(0,6):
    x = np.loadtxt(dir + str(i) + '/outputs_v1/C_dens.dat')[::step,0]
    x = [i+0.1 for i in x]
    y_1 = np.loadtxt(dir + str(i) + '/outputs_v1/C_dens.dat')[::step,1]
    y_2 = np.loadtxt(dir + str(i) + '/outputs_v2/C_dens.dat')[::step,1]
    #output = y_1
    output = [(y_1[i] + y_2[i])/2 for i in range(len(y_1))]
    dat_BLYP[counter] = np.array([x, output]).T
    counter += 1
dir = './intermediate_pressures/'
for i in range(0,6):
    x = np.loadtxt(dir + str(i) + '/outputs_v1/C_dens.dat')[::step,0]
    x = [i+0.1 for i in x]
    y_1 = np.loadtxt(dir + str(i) + '/outputs_v1/C_dens.dat')[::step,1]
    #output = y_1
    y_2 = np.loadtxt(dir + str(i) + '/outputs_v2/C_dens.dat')[::step,1]
    output = [(y_1[i] + y_2[i])/2 for i in range(len(y_1))]
    dat_BLYP[counter] = np.array([x, output]).T
    counter += 1
dir = './high_pressures/'
for i in range(1,6):
    x = np.loadtxt(dir + str(i) + '/outputs_v1/C_dens.dat')[::step,0]
    x = [i+0.1 for i in x]
    y_1 = np.loadtxt(dir + str(i) + '/outputs_v1/C_dens.dat')[::step,1]
    #output = y_1
    y_2 = np.loadtxt(dir + str(i) + '/outputs_v2/C_dens.dat')[::step,1]
    output = [(y_1[i] + y_2[i])/2 for i in range(len(y_1))]
    dat_BLYP[counter] = np.array([x, output]).T
    counter += 1
# ---------------------------------------------------------------------


# Density - Water
dat_BLYP_water = dict()
counter = 1
step = 1
# ---------------------------------------------------------------------
dir = './low_pressures/'
for i in range(0,6):
    x = np.loadtxt(dir + str(i) + '/outputs_v1/OW_dens.dat')[::step,0]
    x = [i+0.1 for i in x]
    y_1 = np.loadtxt(dir + str(i) + '/outputs_v1/OW_dens.dat')[::step,1]
    #output = y_1
    y_2 = np.loadtxt(dir + str(i) + '/outputs_v2/OW_dens.dat')[::step,1]
    output = [(y_1[i] + y_2[i])/2 for i in range(len(y_1))]
    dat_BLYP_water[counter] = np.array([x, output]).T
    counter += 1
dir = './intermediate_pressures/'
for i in range(0,6):
    x = np.loadtxt(dir + str(i) + '/outputs_v1/OW_dens.dat')[::step,0]
    x = [i+0.1 for i in x]
    y_1 = np.loadtxt(dir + str(i) + '/outputs_v1/OW_dens.dat')[::step,1]
    #output = y_1
    y_2 = np.loadtxt(dir + str(i) + '/outputs_v2/OW_dens.dat')[::step,1]
    output = [(y_1[i] + y_2[i])/2 for i in range(len(y_1))]
    dat_BLYP_water[counter] = np.array([x, output]).T
    counter += 1
dir = './high_pressures/'
for i in range(1,6):
    x = np.loadtxt(dir + str(i) + '/outputs_v1/OW_dens.dat')[::step,0]
    x = [i+0.1 for i in x]
    y_1 = np.loadtxt(dir + str(i) + '/outputs_v1/OW_dens.dat')[::step,1]
    #output = y_1
    y_2 = np.loadtxt(dir + str(i) + '/outputs_v2/OW_dens.dat')[::step,1]
    output = [(y_1[i] + y_2[i])/2 for i in range(len(y_1))]
    dat_BLYP_water[counter] = np.array([x, output]).T
    counter += 1
# ---------------------------------------------------------------------

# obtain average water density plots
# ---------------------------------------------------------------------
avg_out = []
max_out = []
min_out = []
err_out = []
for i in range(len(dat_BLYP_water[1])):
    out = []
    for j in range(1,18):
        out.append(np.mean(dat_BLYP_water[j][i,1]))
    avg_out.append(np.average(out))
    max_out.append(np.max(out))
    min_out.append(np.min(out))
    err_out.append(np.std(out))
print(avg_out)

pressures = [1, 5, 10, 20, 30, 40, 60, 65, 70, 75, 80, 85, 100, 200, 300, 400, 500]
```

```python

classical_press = [1,10,20,30,40,60,65,70,75,80,85,100,200,300,400,500]

data_classical_water = dict()
data_classical_co2 = dict()
# ---------------------------------------------------------------------
for i in classical_press:
    dir = f'./classical/MAIN/{i}_atm/'
    x = np.loadtxt(dir + 'OW_dens.dat')[::step,0]
    y = np.loadtxt(dir + 'OW_dens.dat')[::step,1]
    data_classical_water[i] = np.array([x, y]).T

    x = np.loadtxt(dir + 'C_dens.dat')[::step,0]
    y = np.loadtxt(dir + 'C_dens.dat')[::step,1]
    data_classical_co2[i] = np.array([x, y]).T
# ---------------------------------------------------------------------
```

```python
fig, ax = plt.subplots(figsize=(6,6))
plt.rcParams["font.family"] = "Arial"
# ---------------------------------------------------------------------
no_files = 16
def colorFader(c1,c2,mix=0): #fade (linear interpolate) from color c1 (at mix=0) to c2 (mix=1)
    c1=np.array(mpl.colors.to_rgb(c1))
    c2=np.array(mpl.colors.to_rgb(c2))
    return mpl.colors.to_hex((1-mix)*c1 + mix*c2)
# ---------------------------------------------------------------------

# plot of the water density
# ---------------------------------------------------------------------
c1='salmon'
c2='maroon'
n = no_files - 1

counter = 0
for i in classical_press[1:]:
    ax.plot(data_classical_water[i][:, 0], data_classical_water[i][:, 1], '--',
        color=colorFader(c1, c2, counter/n),alpha=( 0.5 + 0.5*(i/500)),
        #label=i,
        linewidth=1.2)
    counter += 1
    
# -------------------------------------------------------------------

# plot of the CO2 density
# ---------------------------------------------------------------------
c1='cyan'
c2='blue'

counter = 0
for i in classical_press:
    ax.plot(data_classical_co2[i][:, 0], data_classical_co2[i][:, 1], '-.',
            color=colorFader(c1, c2, counter/n),linewidth=1.2)
    ax.annotate(xy=(data_classical_co2[i][-1, 0], data_classical_co2[i][-1, 1]), xytext=(6,0),
                textcoords='offset points', text=i, va='center',
                size=9, color=colorFader(c1, c2, counter/n))
    
    counter += 1
    
# ---------------------------------------------------------------------

# ---------------------------------------------------------------------
ax.set_ylabel('Density  (g/ml)', size=14)
ax.set_xlabel(r'Distance  $(\mathrm{\AA})$', size=14)
ax.tick_params(axis="x", which='major', direction="in", labelsize=12,size=5,width=1)
ax.tick_params(axis="y", which='major', direction="in", labelsize=12,size=5,width=1)
ax.tick_params(axis="x", which='minor', direction="in", labelsize=12)
ax.tick_params(axis="y", which='minor', direction="in", labelsize=12)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.set_xlim(-15, 15)
ax.set_ylim(0,1.9)
# ---------------------------------------------------------------------
handles, labels = ax.get_legend_handles_labels()
ax.grid(ls='--')
ax.legend(fontsize=12, loc='upper right')
plt.savefig('./RESULTS/density_classical.png',dpi=400,bbox_inches='tight',facecolor=fig.get_facecolor(), edgecolor='none')
plt.show()
```

```python
bulk_dens_h2o = []
bulk_dens_h2o_err = []
bulk_dens_co2 = []
bulk_dens_co2_err = []

cutoff = 5.0

for i in classical_press[1:]:

    sol = []
    for j in range(len(data_classical_water[i])):
        if data_classical_water[i][j,0] <= -cutoff:
            sol.append(data_classical_water[i][j,1])
    bulk_dens_h2o.append(np.average(sol))
    bulk_dens_h2o_err.append(np.std(sol)/np.sqrt(len(sol)))

    sol = []
    for j in range(len(data_classical_co2[i])):
        if data_classical_co2[i][j,0] <= -cutoff:
            sol.append(data_classical_co2[i][j,1])
    bulk_dens_co2.append(np.average(sol))
    bulk_dens_co2_err.append(np.std(sol)/np.sqrt(len(sol)))

Mr_h2o = 18.01528
Mr_co2 = 44.01
```

```python
bulk_mol_h2o = [i/Mr_h2o for i in bulk_dens_h2o]
bulk_mol_co2 = [i/Mr_co2 for i in bulk_dens_co2]

sol_val_perc = [  (bulk_mol_co2[i]/(bulk_mol_h2o[i] + bulk_mol_co2[i]))*100 for i in range(len(bulk_mol_h2o))]
sol_val_molkg = [ bulk_mol_co2[i]/(bulk_dens_h2o[i]/1000) for i in range(len(bulk_mol_h2o))]  

# ---------------------------------------------------------------------

bulk_mol_h2o_err = [i/Mr_h2o for i in bulk_dens_h2o_err]    
bulk_mol_co2_err = [i/Mr_co2 for i in bulk_dens_co2_err]

sol_val_perc_err = [  (bulk_mol_co2_err[i]/(bulk_mol_h2o[i] + bulk_mol_co2[i]))*100 for i in range(len(bulk_mol_h2o_err))]
sol_val_molkg_err = [ bulk_mol_co2_err[i]/(bulk_dens_h2o[i]/1000) for i in range(len(bulk_mol_h2o_err))]

print(classical_press[1:])
print(sol_val_molkg)
```
